```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("mental_health_workplace_survey.csv")
df.head()
```

```python
print("Total Rows:", df.shape[0])
print("Total Columns:", df.shape[1])
# Check data types and non-null counts
df.info()
# Check summary statistics for numeric columns
df.describe()
```

```python
missing_values = df.isnull().sum()
print("The number of values missing per column are:",missing_values)
```

```python
def remove_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    original_size = df.shape[0]
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    removed = original_size - df.shape[0]
    print(f"{removed} outliers removed from '{col}'")
    return df

for col in ['WorkHoursPerWeek', 'SleepHours']:
    df = remove_outliers_iqr(df, col)
```

```python
sns.scatterplot(x='SleepHours', y='StressLevel', hue='BurnoutRisk', data=df, palette='coolwarm')
plt.xlabel('Stress Level')
plt.ylabel('Sleep Hours')
plt.title('Stress Level vs Sleep Hours')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(x='Age', y='BurnoutLevel', hue='BurnoutRisk', data=df, palette='coolwarm')
plt.xlabel('Age')
plt.ylabel('BurnoutLevel')
plt.title('Age vs BurnoutLevel')
plt.grid(True)
plt.show()
```
